```python
import pandas as pd

#Load the cleaned dataset
cleaned_dataset = pd.read_csv("C:\\Users\\victo\\Downloads\\Renewables\\cleaned_renewables.csv")

#Load the original dataset with the United States
original_dataset = pd.read_csv("C:\\Users\\victo\\Downloads\\Renewables\\sustainable-energy.csv")

#Check if the United States is in the cleaned dataset
if 'United States' not in cleaned_dataset['Entity'].unique():
    print("The United States is not in the cleaned dataset.")
    
    #Extract data for the United States from the original dataset
    us_data = original_dataset[original_dataset['Entity'] == 'United States']

    #Append the United States data to the cleaned dataset
    cleaned_dataset = cleaned_dataset.append(us_data, ignore_index=True)
    
    #Save the updated dataset
    cleaned_dataset.to_csv("C:\\Users\\victo\\Downloads\\Renewables\\updated_cleaned_renewables.csv", index=False)
    print("Updated dataset saved.")
else:
    print("The United States is already in the cleaned dataset.")



```
