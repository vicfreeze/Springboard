```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import the new dataset
output_file_path = "C:\\Users\\victo\\Downloads\\Renewables\\updated_cleaned_renewables.csv"
dataset = pd.read_csv(output_file_path)
# Objective 1: Identify the current status of low-carbon electricity consumption in the United States.
# Hypothesis 1: There is a positive correlation between financial flows to developing countries and renewable energy share.

# Objective 2: Explore the relationship between financial flows and renewable energy adoption.
# Hypothesis 2: Countries with higher GDP growth are more likely to achieve a significant increase 
# in low-carbon electricity consumption.

# Objective 3: Propose strategies to achieve a 15% increase in low-carbon electricity consumption by 2028.
# Hypothesis 3: The implementation of supportive policy frameworks is positively correlated 
# with renewable energy utilization.
```

```python
#Check data types
print(dataset.info())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Filter the dataset for the United States
us_data = dataset[dataset['Entity'] == 'United States']
#Exclude 'name' and 'count' row from summary statistics
summary_stats = us_data['Low-carbon electricity (% electricity)'].describe()
summary_stats.drop(['count'], inplace=True)
#modified summary statistics
print("Summary Statistics for Low-carbon Electricity Consumption in the United States:")
print(summary_stats)
#Low-carbon Electricity Consumption Over Time for the United States
sns.lineplot(x='Year', y='Low-carbon electricity (% electricity)', data=us_data)
plt.title('Low-carbon Electricity Consumption in the United States Over Time')
plt.xlabel('Year')
plt.ylabel('Low-carbon Electricity (%)')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#Renewable Energy Share Across Countries
sns.barplot(x='Entity', y='Renewable energy share in the total final energy consumption (%)', data=dataset)
plt.title('Renewable Energy Share Across Countries')
plt.xlabel('Entity')
plt.ylabel('Renewable Energy Share (%)')
plt.xticks(rotation=45, ha='right')  # Adjust for better readability of country names
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#Scatter Plot: Correlation between Financial Flows and Renewable Energy Share
sns.scatterplot(x='Financial flows to developing countries (US $)', 
y='Renewable energy share in the total final energy consumption (%)', data=dataset)
plt.title('Correlation between Financial Flows and Renewable Energy Share')
plt.xlabel('Financial Flows to Developing Countries (US $)')
plt.ylabel('Renewable Energy Share (%)')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Relationship between GDP Growth, Financial Flows, and Renewable Energy Consumption
sns.scatterplot(x='gdp_growth', y='Renewable energy share in the total final energy consumption (%)', hue='Financial flows to developing countries (US $)', data=dataset)
plt.title('Relationship between GDP Growth, Financial Flows, and Renewable Energy Consumption')
plt.xlabel('GDP Growth')
plt.ylabel('Renewable Energy Share (%)')
plt.legend(title='Financial Flows (US $)')
plt.show()
correlation_matrix = dataset[['gdp_growth', 'Financial flows to developing countries (US $)', 
'Renewable energy share in the total final energy consumption (%)']].corr()
print(correlation_matrix)

#There is a very weak negative correlation between GDP growth and financial flows to developing countries.
#There is a very weak positive correlation between GDP growth and the 
#renewable energy share in the total final energy consumption.
#There is a very weak positive correlation between financial flows to 
#developing countries and the renewable energy share in the total final energy consumption.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#Renewable Energy Share Over Time, purpose is to check trends over time
sns.lineplot(x='Year', y='Renewable energy share in the total final energy consumption (%)', 
data=us_data, label='Renewable Energy Share')
plt.title('Renewable Energy Share Over Time in the United States')
plt.xlabel('Year')
plt.ylabel('Renewable Energy Share (%)')
plt.legend()
plt.show()



```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#Electricity from Fossil Fuels Over Time, purpose is to check trends over time
sns.lineplot(x='Year', y='Electricity from fossil fuels (TWh)', data=us_data, label='Electricity from Fossil Fuels')
plt.title('Electricity from Fossil Fuels Over Time in the United States')
plt.xlabel('Year')
plt.ylabel('Electricity from Fossil Fuels (TWh)')
plt.legend()
plt.show()

```
